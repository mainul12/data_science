## Task: creating a scatterplot

Source: Python Data Analysis (2021, Packt). Page 139

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import random
```

```python
x = random.sample(range(-10, 20), 30)
y = random.sample(range(-50, 20), 30)
```

```python
plt.scatter(x,y,c="orange")
plt.xlabel("Independent variable")
plt.ylabel("Dependent variable")
plt.title("The simplest of scatterplots")
```
