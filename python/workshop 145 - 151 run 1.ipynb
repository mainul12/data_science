## Ticks

```python
import numpy as np
plt.figure(figsize=(6, 3))
plt.plot([2, 4, 6, 8], 'o', [1, 5, 9, 13], 's')
#plt.xticks(ticks=np.arange(4))
```

```python
plt.figure(figsize=(6, 3))
plt.plot([2, 4, 6, 8], 'o', [1, 5, 9, 13], 's')
plt.xticks(ticks=np.arange(4))
```

### Specify tick labels

```python
plt.plot([2, 4, 6, 8], 'o')
plt.xticks(np.arange(4), \
              labels=["Jan", "Feb", "Mar", "Apr"])
plt.xlabel("Month")
plt.ylabel("Y")
```

```python
plt.plot([1, 2, 4, 5], [1, 3, 4, 3], '-o')
plt.xticks(np.arange(1, 6, 2))
plt.xlabel("X")
plt.ylabel("Y")
```
