```python
from symbulate import *
%matplotlib inline
```

```python
from statistics import mean
```

```python
#1000 people 2 flips each

P = BoxModel([1, 0], size = 2)
Y = RV(P, mean)

Y.sim(1000).plot()
```

```python
P = BoxModel([1, 0], size = 2)
Y = RV(P, sum)/2

Y.sim(1000).plot()
```

```python
#1000 people 10 flips each
#the random variable is proportion of heads here (Kottemann, Page 20)

P = BoxModel([1, 0], size = 10)
Y = RV(P, sum)/10

Y.sim(1000).plot()
```

```python
#1000 people 100 flips each
#the random variable is proportion of heads here (Kottemann, Page 20)

P = BoxModel([1, 0], size = 100)
Y = RV(P, sum)/100

Y.sim(1000).plot()
```
