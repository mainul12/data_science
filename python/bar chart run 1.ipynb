```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### The simplest bar chart

```python
plt.style.use("fivethirtyeight")
```

```python
fig, ax = plt.subplots()
ax.bar(["A", "B", "C", "D"], [10, 30, 5, 15])
```

```python
fig, ax = plt.subplots()
ax.bar(["A", "B", "C", "D"], [10, 30, 5, 15], width = 0.4, align="edge")
ax.bar(["A", "B", "C", "D"], [10, 30, 5, 15], width = 0.1)
```

### :D :D :D

```python
fig, ax = plt.subplots()
ax.bar(["A", "B", "C", "D"], [10, 30, 5, 15], width = 0.35, label = "Group 1")
ax.bar(["A", "B", "C", "D"], [40, 30, 15, 5], bottom = [10+0.5, 30+0.5, 5+0.5, 15+0.5], width = 0.50, label = "Group 2")
ax.legend()
```
