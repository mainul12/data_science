### Task: simple demo with multiple subplots 

[source](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplot.html#multiple-subplots)

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
```

```python
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)
fig.suptitle("A tale of 2 subplots")

ax1.plot(x1, y1, "o-")
ax1.set_ylabel("Damped")
ax2.plot(x2, y2, ".-")
ax2.set_ylabel("Undamped")
ax2.set_xlabel("time(s)")
```
