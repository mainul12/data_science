# Parallel Coordinates
If you data has more than 10 features, parallel coordinates might **not** do it for you.To graph them with Pandas and MatPlotLib, you have to specify a feature to group by (it can be non-numeric). 

```python
from sklearn.datasets import load_iris
```

```python
from pandas.tools.plotting import parallel_coordinates
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use("ggplot")
%matplotlib inline
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
df.head(3)
```

### Adding a new column "target_names"

```python
df["target_names"] = [data.target_names[i] for i in data.target]
```

```python
df.head()
```

```python
plt.figure(figsize=(10,8))
parallel_coordinates(df,"target_names")
plt.show()
```
